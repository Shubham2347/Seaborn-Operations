```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distribution within Categories
```

```python
"""
Boxplot
Violinplot
Swarmplot
Boxenplot
"""
```

```python
df= pd.read_csv("StudentsPerformance.csv")
df
```

```python
df.head()
```

# boxplot

```python
plt.figure(figsize=(10,4),dpi=300)
sns.boxplot(data=df,y="math score",x="test preparation course")
```

```python
plt.figure(figsize=(10,8),dpi=300)
sns.boxplot(data=df,x="reading score",y="parental level of education",
           hue="test preparation course",palette="Set2")
plt.legend(bbox_to_anchor=(1,0.5))
```

# violinplot

```python
plt.figure(figsize=(10,8),dpi=300)
sns.violinplot(data=df,x="reading score",y="parental level of education"
           )
#plt.legend(bbox_to_anchor=(1,0.5))
```

```python
plt.figure(figsize=(10,8),dpi=300)
sns.violinplot(data=df,x="reading score",y="parental level of education",
           hue="test preparation course",palette="Set2",
              split=True,inner="stick")#inner="quartile"
plt.legend(bbox_to_anchor=(1,0.5))
```

```python
plt.figure(figsize=(10,8),dpi=400)
sns.violinplot(data=df,x="reading score",y="parental level of education",
         bw=0.1)
#plt.legend(bbox_to_anchor=(1,0.5))
```

# swarmplot

```python
sns.swarmplot(data=df,x="reading score",size=2,y="gender",
             hue="test preparation course",dodge=True)
plt.legend(bbox_to_anchor=(1.35,0.5))
```

# boxenplot

```python
sns.boxenplot(x="math score",y="test preparation course",data=df,hue="gender")
```
