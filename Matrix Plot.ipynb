```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("country_table.csv")
```

```python
df=df.set_index("Countries")
```

```python
df
```

# heatmap

```python
plt.figure(dpi=300)
sns.heatmap(df.drop("Life expectancy",axis=1),linewidth=0.5,
            annot=True)
```

```python
plt.figure(dpi=300)
sns.heatmap(df.drop("Life expectancy",axis=1),linewidth=0.5,
            annot=True,cmap='viridis')
```

# clustermap

```python
plt.figure(dpi=300)
sns.clustermap(df.drop("Life expectancy",axis=1),linewidth=0.5,
            annot=True,cmap='viridis')
```
